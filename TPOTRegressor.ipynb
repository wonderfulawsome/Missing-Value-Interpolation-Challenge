```python
import pandas as pd
data = pd.read_csv('C:/Users/82106/Desktop/데이터 분석 프로젝트 2/dacon 결측치 보간/open/data.csv')
```

```python
from tpot import TPOTRegressor
from sklearn.model_selection import train_test_split


# 결측치가 있는 행과 없는 행을 분리
train = data.dropna(subset=['Value'])
test = data[data['Value'].isnull()]

# 독립 변수와 종속 변수 분리 (예제에서는 'id'에서 숫자만 추출)
train['Time'] = pd.to_numeric(train['id'].str.extract('(\d+)')[0])
test['Time'] = pd.to_numeric(test['id'].str.extract('(\d+)')[0])
X_train = train[['Time']]
y_train = train['Value']
X_test = test[['Time']]

# TPOT 회귀 모델 생성
tpot = TPOTRegressor(generations=5, population_size=50, verbosity=2, random_state=42)
tpot.fit(X_train, y_train)
```

```python
# 결측치 예측
predicted_values = tpot.predict(X_test)
test['Value'] = predicted_values
```

```python
# 예측 결과 확인
print(test[['id', 'Value']])
```

```python
# 예측된 값을 원래 데이터셋에 적용 (결측치 대체)
data.loc[data['Value'].isnull(), 'Value'] = test['Value']
```

```python
submission = pd.read_csv('C:/Users/82106/Desktop/데이터 분석 프로젝트 2/dacon 결측치 보간/open/sample_submission.csv')
submission['Value'] = data['Value']
```
