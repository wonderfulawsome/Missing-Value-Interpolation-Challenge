```python
import pandas as pd
import matplotlib.pyplot as plt

# 결측치 유무 확인
data['Missing'] = data['Value'].isnull()

# 결측치 분포 시각화
plt.figure(figsize=(15, 6))
plt.plot(data['id'], data['Missing'], marker='o', linestyle='None', markersize=2)
plt.title('Missing Value Distribution Over IDs')
plt.xlabel('ID')
plt.ylabel('Missing (True/False)')
plt.yticks([0, 1], ['Not Missing', 'Missing'])
plt.grid(True)
plt.show()
```
